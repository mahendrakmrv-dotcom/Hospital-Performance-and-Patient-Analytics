## Converting object to int to perform mean, on missing values

```python
df['wait_time_min'] = pd.to_numeric(df['wait_time_min'], errors='coerce').astype(np.int64)
```

```python
df['wait_time_min'].fillna(df['wait_time_min'].mean(),inplace=True)
```

```python
print(df.dtypes)
```

```python
df['wait_time_min'].isnull().sum()
```

```python
df['wait_time_min'] = df['wait_time_min'].replace(0, np.nan)
```

```python
df['wait_time_min'] = df['wait_time_min'].fillna(df['wait_time_min'].mean())
```

```python
df['treatment_cost'] = pd.to_numeric(df['treatment_cost'], errors='coerce').fillna(0).astype(np.int64)
```

```python
df['treatment_cost'] = df['treatment_cost'].replace(0, np.nan)
```

```python
df['treatment_cost'] = df['treatment_cost'].fillna(df['treatment_cost'].mean())
```

```python
def fix(x):
    if x in ['1','Yes','TRUE']:
        return 'Yes'
    else:
        return 'No'
```
